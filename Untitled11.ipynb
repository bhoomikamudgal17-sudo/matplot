```python
df=pd.read_csv('batter.csv')
```

```python
df.head()
```

```python
sample=df.head(100).sample(25,random_state=5)
```

```python
sample
```

```python
plt.figure(figsize=(15,7))
plt.scatter(sample['avg'],sample['strike_rate'],s=sample['runs'])
plt.xlabel('avg')
plt.ylabel('strike_rate')
plt.title('avg vs strike rate')
plt.axhline(130,color='red',linestyle='dashed')
plt.axvline(30,color='red',linestyle='dashed')
for i in range(sample.shape[0]):
    plt.text(sample['avg'].values[i],sample['strike_rate'].values[i],sample['batter'].values[i])
plt.show()
```

```python
#subplots
```

```python
fig,ax=plt.subplots()
ax.scatter(df['avg'],df['strike_rate'])
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')
ax.set_title('avg vs strike rate')
```

```python
fig, ax=plt.subplots(nrows=2,ncols=1)
```

```python
fig, ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].scatter(df['avg'],df['strike_rate'])
ax[0,0].set_xlabel('avg')
ax[0,0].set_ylabel('strike_rate')
ax[0,0].set_title('avg vs strike rate')
ax[1,0].scatter(df['avg'],df['strike_rate'])
ax[1,0].set_xlabel('avg')
ax[1,0].set_ylabel('strike_rate')
ax[1,0].set_title('avg vs strike rate')
ax[1,1].hist(df['avg'])
ax[0,1].hist(df['runs'])
ax[1,1].set_xlabel('avg')
ax[1,1].set_ylabel('strike_rate')
ax[1,0].set_title('avg vs strike rate')
plt.tight_layout() #=>to avoid overlap
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
```

```python
z.shape
```

```python
fig=plt.figure(figsize=(10,7))
ax=plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
plt.show()
```

```python
#contour plot 3d to 2d
```

```python
fig=plt.figure(figsize=(10,7))
ax=plt.subplot()
ax.contourf(xx,yy,z,cmap='viridis')
plt.show()
```
